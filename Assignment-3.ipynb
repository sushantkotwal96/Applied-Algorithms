### Add multiplication to the Python class overload asterisk.


```python
class complex_:
    def __init__(self, re=0, im=0):
        self.re = re
        self.im = im

    def get_re(self):
        return self.re

    def get_im(self):
        return self.im

    def __str__(self):
        g = lambda x:"+" if x >= 0 else ""
        return f"({self.re}{g(self.im)}{self.im}i)"

    def cadd(self, other):
        new_re = self.get_re() + other.get_re()
        new_im = self.get_im() + other.get_im()
        return complex_(new_re, new_im)

    def __add__(self, other):
        new_re = self.get_re() + other.get_re()
        new_im = self.get_im() + other.get_im()
        return complex_(new_re, new_im)


    def __mul__(self, other):
        new_re = (self.get_re()*other.get_re()) - (self.get_im()*other.get_im())
        new_im = (self.get_re()*other.get_im()) + (other.get_re()*self.get_im())
        return complex_(new_re, new_im)

if __name__ == '__main__':

    w = complex_(1,-3)
    x = complex_(-1,3)
    y = complex_(1,3)
    z = complex_(-1,-3)
    print(w)
    print(x)
    print(y)
    print(z)

    print(w.cadd(x).cadd(y).cadd(z))

    print((1-3j) + (-1+3j) + (1+3j) + (-1-3j))

    print(w + x + y + z)


    print(w*x*y*z)

    print((1-3j) * (-1+3j) * (1+3j) * (-1-3j))
```

## Solution:
### 1. In the above program, we are implementing Operator Overloading. 
### 2. The default functioning of '+' operator is changed according to formula for complex numbers using \_\_add\_\_method. 
### 3. Likewise for overloading the ' * ' operator i.e the multiplication, we change the \_\_mul\_\_() function in the class Complex as per the formula of multiplication of complex numbers 
### 4. We are defining the __mul__() function and giving the formula for multiplication of complex numbers there. ### 5. So when we use the * operation for statement w * x * y * z, the formula in \_\_mul\_\_() is executed i.e. the default multiplication is overriden and we get our output accordingly.

## Problem 5
### In this question, you will work with the three algorithms, algorithm1, algorithm2, algorithm3, which check whether a given sequence is unique. Perform an experimental analysis to determine the largest value of input size such that the given algorithm runs in less than 45 seconds.

### Algorithm 1

```python
import time
from timeit import default_timer as timer
def algorithm1( S):
#S:sequence
    for j in range(len(S)):
        for k in range(j+1, len(S)):
            if S[j] == S[k]:
                return False
    return True

if __name__ == '__main__':
    run_time=0
    i = 1000
    l1=[]
    rt1=[]
    while run_time<=45:
        list1=list(range(i))
        start_time=time.time()
        algorithm1(list1)
        run_time=time.time()-start_time
        l1.append(i)
        rt1.append(run_time)
        print("The runtime for input size {} is {}".format(i,run_time))
        #l1[i]=running_time
        i+=2000
    #print(l1)
    print("The maximum no. of inputs for algorithm 1 such as is executes within 45 seconds is: ~",l1[-2])
```

## Solution:
### 1. In the above algorithm, we are checking the maximum input size such that the algorithm runs within 45 seconds.
### 2. We are creating and passing a list of varying size to the algorithm and checking runtime for each input size.
### 3. For doing this, we have implemented a while loop which starts from 1000 and we are incrementing the number of input size by 2000 after each iteration.
### 4. We are checking the condition whether the run_time is less than or equal to 45. If this condition is not satisfied the loop breaks.
### 5. This maximum input size which we are getting can vary from one system to other.
### 6. We can say that this algorithm has a time complexity of O(n^2)

## Plot for Algorithm1

```python
from matplotlib import pyplot as plt
plt.xlabel("Inputs size")
plt.ylabel("Runtime")
plt.plot(l1,rt1, marker ="o",label='Algorithm 1 plot')
plt.legend()
plt.show()
```

```python
import time
from timeit import default_timer as timer
def algorithm2(S):
    #S:sequence
    S = sorted(S)
    for j in range(1, len(S)):
        if S[j-1] == S[j]:
          return False
    return True

if __name__ == '__main__':
    run_time=0
    i = 30000000
    l2=[]
    rt2=[]
    while run_time<=45:
        list2=list(range(i))
        start_time=time.time()
        algorithm2(list2)
        run_time=time.time()-start_time
        l2.append(i)
        rt2.append(run_time)
        print("The runtime for input size {} is {}".format(i,run_time))
        #l1[i]=running_time
        i+=15000000
    #print(l1)
    print("The maximum no. of inputs for algorithm 2 such as is executes within 45 seconds is: ~",l2[-2])
```

### Algorithm 3

```python
import time
from timeit import default_timer as timer
def algorithm3(S, start, stop):
#slice S[start:stop], S:sequence
    if stop - start <= 1: return True
    elif not algorithm3(S, start, stop-1): return False
    elif not algorithm3(S, start+1, stop): return False
    else: return S[start] != S[stop-1]
    return True

if __name__ == '__main__':
    run_time=0
    i = 22
    l3=[]
    rt3=[]
    while run_time<=45:
        list3=list(range(i))
        start_time=time.time()
        algorithm3(list3,0,i)
        run_time=time.time()-start_time
        l3.append(i)
        rt3.append(run_time)
        print("The runtime for input size {} is {}".format(i,run_time))
        #l1[i]=running_time
        i+=1
    #print(l1)
    print("The maximum no. of inputs for algorithm 3 such as is executes within 45 seconds is: ~",l3[-2])
```
